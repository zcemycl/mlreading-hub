```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import beta,uniform,norm
```

The **Acceptance-Rejection** method is an algorithm to generate random variates from a distribution for which it may be difficult to sample from. Let $X$ and $Y$ be two random variables with respective PDF f(hard) and g(proposal), and let c be a constant such that, for all $x$, we have,

$$f(x)\leq cg(x)$$

For the univariate case, the algorithm proceeds as follows:

1. Generate $x$ from a Proposal distribution, easy to sample from.
2. Generate $y$ uniformly in the interval $[0,cg(x)]$ 

$$c\leq \max f(x)$$
3. Accept $x$ if $y\leq f(x)$ else repeat.

---

Let $Y\sim f_Y(y)$ and $V\sim f_V(V)$, where $f_Y$ and $f_V$ have common support with, 

$$M = \sup_y f_Y(y)/f_V(y) < \infty \rightarrow\frac{f_Y(y)}{M f_V(y)}\leq 1$$

where $\sup$ is supremum, it is kind of like maximum value, but it is the smallest number that is $\geq$ to every number in the set. For example, the interval $[0,1)$ has no maximum value, but 1 is the supremum of the interval.

To generate a random variable $Y\sim f_Y$,

a. Generate $U \sim \mathbb{U}(0,1)$, $V\sim f_V$, independent. 

b. If $U<\frac{1}{M}f_Y(V)/f_V(V)$, set $Y=V$; otherwise, return to step (a).

#### My personal notes
1. Make sure the intervals of target and proposal distributions matches. 
2. The constant is to make area under proposal distribution should be greater than that under the target, and the maximum value of the target should be covered under proposal curve. Therefore, 
$$c\leq \max\bigg\{\frac{f(x)}{g(x)}\bigg\}$$
3. To enhance the acceptance ratio, choose the proposal distribution which should not have too much coverage area.
4. The distribution $U\sim \mathbb{U}(0,1)$ is to accept or reject the sample.

**Example 1** Sample from beta distribution $B(6.3,2.7)$. Use the Acceptance-Rejection algorithm to sample from this distribution using $N=10000$ simulations.

$$\begin{align*}
g(x) &\sim \mathbb{U}[0,1] \\
f(x) &\sim B(6.3,2.7)\\
\end{align*}$$

```python
N = 10000
X = np.random.uniform(0,1,N)
c = max(beta.pdf(X,6.3,2.7))
# c=1
```

Since $c\leq \max f(x)$, $c=2.67$

```python
Y = np.random.uniform(0,1,N)
# plt.hist(X[Y<=beta.pdf(X,6.3,2.7)],40)
accept = Y[X<=beta.pdf(Y,6.3,2.7)/c]
# plt.scatter(accept,beta.pdf(accept,6.3,2.7))
plt.hist(accept,40,alpha=0.2)
```

```python
sa = X<=beta.pdf(Y,6.3,2.7)/c
sr = X>beta.pdf(Y,6.3,2.7)/c
plt.scatter(Y[sa],X[sa],marker="x",label="accept")
plt.scatter(Y[sr],X[sr],marker="x",label="reject")
plt.legend()
```

```python
# Acceptance ratio
len(Y[sa])/len(X)
```

Without c, it will not work.

```python
sa = X<=beta.pdf(Y,6.3,2.7)
sr = X>beta.pdf(Y,6.3,2.7)
plt.scatter(Y[sa],X[sa],marker="x",label="accept")
plt.scatter(Y[sr],X[sr],marker="x",label="reject")
plt.legend()
```

```python
# Acceptance ratio
len(Y[sa])/len(X)
```

```python
# estimated mean
print(np.mean(Y[sa]))
# theoretical mean
print(6.3/(6.3+2.7))
```

**Example 2** Normal Distribution

```python
N = 1000
f = lambda x,m,v: np.exp(-(x-m)**2/2/v**2)/np.sqrt(2*np.pi*v**2)
X = np.random.uniform(0,10,N)
c = np.max(f(X,5,1)/uniform.pdf(X,0,10))
print(c)
Y = np.random.uniform(0,10,N)
U = np.random.uniform(0,1,N)
```

```python
plt.scatter(X,c*uniform.pdf(X,0,10),s=0.5)
plt.scatter(X,f(X,5,1),marker='x',s=0.5)
```

```python
sa = U<=f(Y,5,1)/uniform.pdf(Y,0,10)/c
sr = U>f(Y,5,1)/uniform.pdf(Y,0,10)/c
# acceptX = X[sa]
acceptY = Y[sa]
plt.scatter(acceptY,U[sa],s=0.5)
```

```python
plt.hist(acceptY,40,alpha=0.2)
```

```python
plt.scatter(Y[sa],U[sa],marker="x",label="accept")
plt.scatter(Y[sr],U[sr],marker="x",label="reject")
plt.legend()
```

```python
print(np.mean(Y[sa]))
```
