**Random Walk Metropolis** is one of the examples of MCMC algorithms, it is also regarded as dependent sampling because each step we choose depends on the previous step.  It is useful to estimate the posterior distribution which is hard to sample from.

$$\begin{align*}
p(\theta|x)&=\frac{p(x|\theta)p(\theta)}{p(x)} \\
&\propto p(x|\theta)p(\theta)
\end{align*}$$

Algorithm is outlined as follows,
1. Start with, 
    - sample a parameter from a proposal distribution, 
    $$\theta_0 \sim \pi(\theta)$$
    - define a transition distribution with a step size parameter, 
    $$\theta'_t \sim \mathcal{N}(\theta_{t-1},\sigma)$$
    where $\sigma$ is the step size.
    - define likelihood and prior for sampling ratio,
    $$\begin{align*}
    p(X|\theta) &\sim \prod_i\mathcal{N}(x_i|\theta) \\
    p(\theta) &\sim \mathbb{U}
    \end{align*}$$
    where the prior is sometimes not the transition probability $q(\theta'_t|\theta_{t-1})$
2. Iterates over,
    - Draw a new sample from transition distribution,
    $$\theta'_t \sim \mathcal{N}(\theta_{t-1},\sigma)$$
    - Calculate the ratio,
    $$r = \frac{p(X|\theta'_t)p(\theta'_t)}{p(X|\theta_{t-1})p(\theta_{t-1})}$$
    - Compare with random variable $u\sim\mathbb{U}(0,1)$, if $r>u$ accept else reject.
    - Accept:
    $$\theta_t = \theta'_t$$
    - Reject:
    $$\theta_t = \theta_{t-1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
# %matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal,norm
import matplotlib.colors as colors
import matplotlib.cm as cm
```

```python
def sampleGauss2D(mu,cov,xmin=-10,xmax=10,ymin=-10,ymax=10):
    """
    Given x,y mean and covariance,
    Sample 2D Gaussian Distribution Datapoints 
    within x[-10,10], y[-10,10].
    mu = [1,2] w
 x   cov = [2,2]
    """
    x = np.linspace(xmin,xmax,100)
    y = np.linspace(ymin,ymax,100)
    x,y = np.meshgrid(x,y)
    pos = np.empty(x.shape+(2,))
    pos[:,:,0]=x
    pos[:,:,1]=y
    rv = multivariate_normal(mu,cov)
    z = rv.pdf(pos)
    return x,y,z
```

```python
def plot3DSurface2DContour(xs,ys,zs,xmin=-10,xmax=10,ymin=-10,ymax=10):
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(121, projection='3d')
    xs=np.array(xs)
    xs=np.sum(xs,axis=0)
    ys=np.array(ys)
    ys=np.sum(ys,axis=0)
    zs=np.array(zs)
    zs=np.sum(zs,axis=0)
    zmax = np.max(zs)
    ax.set_xlim3d([xmin,xmax])
    ax.set_ylim3d([ymin,ymax])
    ax.set_zlim3d([0,zmax])
    ax.view_init(75, 100)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.plot_surface(xs,ys,zs,cmap="plasma")

    ax = fig.add_subplot(122)
    ax.contour(xs,ys,zs)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')
```

```python
def demo3d():
    x0,y0,z0 = sampleGauss2D([2,2],[[1,0],[0,4]])
    x1,y1,z1 = sampleGauss2D([0,-5],[[2,1.5],[1,2]])
    x2,y2,z2 = sampleGauss2D([-3,-3],[[2,0],[0,2]])
    plot3DSurface2DContour([x0,x1,x2],[y0,y1,y2],[z0,z1,z2])
```

```python
demo3d()
```

```python
def sampleGauss1D(mu,var):
    """
    Given x mean and variance,
    sample 1D Gaussian Distribution Datapoints
    with x[-10,10].
    mu = [1]
    cov = [1]
    """
    x = np.linspace(-10,20,10000)
    rv = norm(mu,var)
    y = rv.pdf(x)
    return x,y
```

```python
def plot2D(xs,ys):
    fig = plt.figure()
    xs = np.array(xs).sum(axis=0)
    ys = np.array(ys).sum(axis=0)
    plt.plot(xs,ys)
    plt.xlabel('x')
    plt.ylabel('y')
```

```python
class experiment:
    def __init__(self,samples):
        self.samples = samples
        self._logL = lambda m,v,m00,v00:(np.log(\
                norm.pdf(samples,m,v))\
                +np.log(norm.pdf(m,m00,1))\
                +np.log(norm.pdf(v,v00,1))).sum()
    def explore(self,m0,v0,szM,szV,T=300000):
        self.ms,self.vs = np.zeros(T),np.zeros(T)
        self.its,self.rs = np.zeros(T),np.zeros(T)
        mt,vt = m0,v0
        for it in tqdm(range(T)):
            mp = szM*np.random.randn(1)+mt
#             vp = szV*np.random.randn(1)+vt
            vp = np.random.gamma(vt,szV,1)
            while vp<=0:
#                 vp = szV*np.random.randn(1)+vt
                vp = np.random.gamma(vt,szV,1)
            logR = self._logL(mp[0],vp[0],mt,vt)-\
                    self._logL(mt,vt,mt,vt)
            u = np.random.rand(1)[0]
            if self._logL(mp[0],vp[0],mt,vt)>self._logL(mt,vt,mt,vt):
                mt,vt = mp[0],vp[0]
            elif np.exp(logR)>u:
                mt,vt = mp[0],vp[0]
            self.ms[it]=mt
            self.vs[it]=vt
            self.rs[it]=np.exp(logR)
            self.its[it]=it
    def mu_var_evolution(self):
        plt.figure(figsize=(20,10))
        ax1 = plt.subplot(1,2,1)
        ax1.plot(self.its,self.ms)
        ax1.set_xlabel('iterations')
        ax1.set_ylabel('mean')

        ax2 = plt.subplot(1,2,2)
        ax2.plot(self.its,self.vs)
        ax2.set_xlabel('iterations')
        ax2.set_ylabel('variance')
    def histPath(self,x0,y0,z0,xmin=-10,xmax=10,
                 ymin=0,ymax=15,bins=50):
        fig = plt.figure(figsize=(20,10))
        ax1 = fig.add_subplot(121, projection='3d')
        x, y = np.array(self.ms),np.array(self.vs)
        hist,xedges,yedges = np.histogram2d(x,y,
                bins=bins, range=[[xmin,xmax],[ymin,ymax]])

        # Construct arrays for the anchor positions of the 16 bars.
        xpos, ypos = np.meshgrid(xedges[:-1]+0.25,
                yedges[:-1]+0.25,indexing="ij")
        xpos = xpos.ravel()
        ypos = ypos.ravel()
        zpos = 0

        # Construct arrays with the dimensions for the 16 bars.
        dx = dy = 0.5 * np.ones_like(zpos)
        dz = hist.ravel()
        offset = dz + np.abs(dz.min())
        fracs = offset.astype(float)/offset.max()
        norm = colors.Normalize(fracs.min(), fracs.max())
        cs = cm.jet(norm(fracs))

        ax1.bar3d(xpos,ypos,zpos,dx,dy,dz,zsort='average',
                  color=cs)
        ax1.view_init(75, 100)

        ax = fig.add_subplot(122)
        ax.contour(x0,y0,z0)
        ax.plot(self.ms,self.vs)
        ax.set_xlim([xmin,xmax])
        ax.set_ylim([ymin,ymax])
        ax.set_xlabel('mean')
        ax.set_ylabel('variance')
        ax.set_aspect('equal')
```
