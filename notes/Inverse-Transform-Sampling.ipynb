```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import norm,uniform
```

**Inverse transform sampling** is a basic method for pseudo-random number sampling, i.e. for generating sample numbers at random from any probability distribution given its cumulative distribution function. It works as follows,
1. Generate a random number $u$ from the standard uniform distribution in the interval $[0,1]$.
$$U\sim \mathbb{U}[0,1]$$
2. Find the inverse of the desired CDF. 
$$F_X^{-1}(x)$$
3. Computer $X=F^{-1}_X(u)$. The computed random variable $X$ has distribution $F_X(x)$.

CDF is always an increasing function, it represents the probability that $X$ will take a value less than or equal to $x$, and is defined as $F_X(x)$. If there is a transformation such that it turns probability to $X$, i.e. $T:[0,1]\rightarrow \mathbb{R}$, so that $T(U)=^d X$ and $U\sim \mathbb{U}[0,1]$.

$$F_X(x) = P(X\leq x)=P(T(U)\leq x) = P(U\leq T^{-1}(x))=T^{-1}(x)$$

Inverse CDF is the probability that $U$ will take a value less than or equal to $F_X(x)=y$.

```python
lamb,N = 1,1000
X = np.linspace(0,10,N)
f = lambda x: np.exp(-x/lamb)/lamb
F = lambda x: (1-np.exp(-x/lamb))
invF = lambda p: -lamb*np.log(1-p)
```

```python
plt.plot(X,f(X))
```

```python
plt.plot(X,F(X))
```

```python
p = np.linspace(0,1,N)
plt.plot(p,invF(p))
```

```python
U = uniform.rvs(0,1,N)
_ = plt.hist(U,bins=50)
```

```python
_ = plt.hist(invF(U),bins=50)
```
