```python
def softmax(vec):
    vec -= vec.max() # avoid overflow
    vec = np.exp(vec)
    return vec/vec.sum(axis=1,keepdims=1)
```

```python
def tanh(vec):
    pos = np.exp(vec)
    neg = np.exp(-vec)
    return (pos-neg)/(pos+neg)
```

```python
def avgLosses(losses,every=1000):
    initSum = sum(losses[:every])
    res = [initSum/every]
    for i in range(every,len(losses)):
        initSum+=losses[i]-losses[i-every]
        res.append(initSum/every)
    return res
```

```python
def forward(W1,W2,sent,eps=1e-12):
    H = XdotW(W1,sent[:-1])
    H = tanh(H)
    pred = softmax(H.dot(W2))
    pred=np.clip(pred,eps,1-eps) # clip interval [eps,1-eps]
    return H,pred
```

#### Bigrams Autoencoder
A model with one hidden layer of less dimensions $k<V$. 

$$\begin{align*}
p(Y|X)&=p(X_{i+1}|X_i)=\sigma(HW_2) \\
&= \sigma(\tanh(XW_1)W_2)
\end{align*}$$

,where $\sigma(z)_i = \frac{e^{z_i}}{\sum^K_{j=1}e^{z_j}}$ a softmax function, $\tanh(x) = \frac{e^x-e^{-x}}{e^x+e^{-x}}$ is a hyperpolic tangent function, $Y$ and $X$ are matrices $[N\times V]$, $W_1$ $[V\times k]$ and $W_2$ $[k\times V]$ matrices. $N$ is the number of words in the sentence, $k$ the dimension of hidden layer, and $V$ the number of vocabularies.

The loss function of a sentence is given by,
$$J=\frac{1}{N}\sum^N_{n=1}\sum^V_{k=1}y_{nk}\log(p(y_{nk}|x_n))$$

The parameters will be updated by gradient descents, and the gradients of weight matrices are given by, 

$$\begin{align*} 
W_2\leftarrow W_2-\eta\nabla_{W_2}J &\Rightarrow  \nabla_{W_2}J = H^\top\big(p(Y|X)-Y\big)\\
W_1\leftarrow W_1-\eta\nabla_{W_1}J &\Rightarrow  \nabla_{W_1}J = X^\top\bigg(\big(p(Y|X)-Y\big)W_2^\top\odot\big(1-H^2\big)\bigg)\\
\end{align*}$$

where $\odot$ the element-wise multiplication operation.

#### Advantages
- It requires less space since matrices are compressed.
- It requires less computation time.

```python
V,k,eps = len(w2id),100,1e-12
W1 = np.random.uniform(0,1,(V,k))
W2 = np.random.uniform(0,1,(k,V))
lr,epochs,losses = 1e-1,300,[]
for e in tqdm(range(epochs)):
    random.shuffle(newSents)
    for sent in newSents:
        L = len(sent)
        inder = np.arange(L-1)
        H,pred=forward(W1,W2,sent,eps)
        loss = -(np.log(pred+eps))[inder,sent[1:]].sum()/L
        pred[inder,sent[1:]]-=1
        
        dW2 = H.T.dot(pred)
        includes = XtransposeDot(sent[:-1])
        tmp = (pred.dot(W2.T))*(1-H**2)
        dW1 = np.zeros((V,k))
        for key,val in includes.items():
            dW1[key] = tmp[val].sum(axis=0)
        
        W1-=lr*dW1
        W2-=lr*dW2
        losses.append(loss)
        
```

```python
avglosses = avgLosses(losses,every=10000)
plt.plot(range(len(avglosses)),avglosses)
```

#### Bigrams Language Models
- The log likelihood is given by,
$$\begin{align*}
p(\{w_i\}^T_{i=1})&=p(w_1)\prod^T_{t=2}p(w_t|w_{t-1})\\
\log p(\{w_i\}^T_{i=1})&=\log p(w_1)+\sum^T_{t=2}\log p(w_t|w_{t-1})\\
\frac{1}{T}\log p(\{w_i\}^T_{i=1})&=\frac{1}{T}\bigg(\log p(w_1)+\sum^T_{t=2}\log p(w_t|w_{t-1})\bigg)
\end{align*}$$

```python
res,resfake = [],[]
for i,sent in enumerate(tqdm(newSents)):
    logprob,logprobfake = 0,0
    L = len(sent)
    inder = np.arange(L-1)
    H,pred=forward(W1,W2,sent,eps)
    logprob=(np.log(pred[inder,sent[1:]])).sum()
    logprob/=L
    res.append(logprob)
    #
    sent_ = [0]+list(np.random.randint(2,
                    len(w2id),L-2))+[1]
    H,pred=forward(W1,W2,sent_,eps)
    logprobfake=(np.log(pred[inder,sent_[1:]])).sum()
    logprobfake/=L
    resfake.append(logprobfake)
    
```
